##  ***Data Wrangling***

```python
# Null Values are replaced using Loop in columns children, agent, compnay
null_col = ['children', 'agent', 'company']
for columns in null_col:
  df[columns].fillna(0, inplace=True)
```

```python
# Null values of column country are replacedd by other using loop
df['country'].fillna('others', inplace=True)
```

#### Adding new Columns

```python
df.columns
```

```python
# Calculating total days staying in Hotel
df['total_stay'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
```

```python
# Calculating total number of People stays in Hotel
df['total_people'] = df['adults'] + df['children'] + df['babies']
```

```python
# Calculating total number of people stays in hotel is zero
(df[df['total_people'] == 0].shape)
```

```python
# Removimg all rows that have total number of people is Zero
df.drop(df[df['total_people'] == 0].index, inplace =True)
```

### What all manipulations have you done and insights you found?

*   We have 4 Columns('company','agent','country','children') that have missing values. These values were replced by '0' & Country by 'others'. 
*   In Dataset I have added 2 Columnsi.e. total_stay & total_people.
*   I have removed the Rows that have 0 person in the column of Total_people

#### Chart - 2

Visulizition of Distribution Channels of Booking

```python
group_by_distribution = df.groupby('market_segment')
df1 = pd.DataFrame(round((group_by_distribution.size()/df.shape[0])*100,2)).reset_index().rename(columns = {0:'Booking_%'})
df1
```

*   For visulization the market segment of booking I choose the pie chart. Because It shows the percentage of total booking by each distribution channel

#### Chart - 3

Month wise Booking

```python
# Arranging the Data Month wise
monthly_booking=df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':"Number of booking"})
sequence_of_months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_booking['arrival_date_month']=pd.Categorical(monthly_booking['arrival_date_month'],categories=sequence_of_months,ordered=True)
monthly_booking=monthly_booking.sort_values('arrival_date_month')
monthly_booking
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=monthly_booking, x="arrival_date_month", y="Number of booking")
plt.title("Number of Bookings in Months", fontweight='bold', fontsize = 25)
plt.xlabel('Month', fontsize = 15)
plt.ylabel('Booking in Month-wise', fontsize = 15, fontweight='bold')
```

*   **Barplot**: Barplot is very easy to display the categorical Data. With the help of Barplot It is very easy to anylisize the month wise booking. It makes comparion very easy.
*  July and August months have most bookings this is positive impact on business. November, December and January months have less bookings this is neative impact.

*   July and August months have bookings above the average bookings and November, December and January months have bookings below the average bookings.

#### Chart - 5


Booking cancellation Analysis

```python
waiting_bookings = df[df['days_in_waiting_list'] !=0]  # Selecting bookings with non zero waiting time
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.kdeplot(ax=axes[0],x = 'days_in_waiting_list', hue = 'is_canceled' , data = waiting_bookings)
sns.kdeplot(ax = axes[1], x = df['lead_time'], hue = df['is_canceled'])
plt.show()
```


*   We see that most of the bookings that are cancelled have waiting period of less 150 days but also most of bookings that are not cancelled also have waiting period less than 150 days. Hence this shows that waiting period has no effect on cancellation of bookings.

*   Also, lead time has no affect on cancellation of bookings, as both curves of cancelation and not cncelation are similar for lead time too.
